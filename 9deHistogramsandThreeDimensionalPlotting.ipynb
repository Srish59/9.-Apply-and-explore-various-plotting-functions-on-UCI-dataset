```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # To visualize
from mpl_toolkits.mplot3d import Axes3D
data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv')
data
data['Glucose'].plot(kind='hist')
```

```python
fig = plt.figure(figsize=(4,4))
ax = fig.add_subplot(111, projection='3d')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = data['Age'].values
y = data['Glucose'].values
z = data['Outcome'].values
ax.set_xlabel("Age (Year)")
ax.set_ylabel("Glucose (Reading)")
ax.set_zlabel("Outcome (0 or 1)")
ax.scatter(x, y, z, c='r', marker='o')
plt.show()
```
