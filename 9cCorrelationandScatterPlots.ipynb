```python
import pandas as pd
con = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv')
con
list(con.columns)
import seaborn as sns
sns.scatterplot(x="Pregnancies", y="Age", data=con)
```

```python
sns.lmplot(x="Pregnancies", y="Age", data=con)
```

```python
sns.lmplot(x="Pregnancies", y="Age", hue="Outcome", data=con);
```

```python
from scipy import stats
stats.pearsonr(con['Age'], con['Outcome'])
```

```python
cormat = con.corr()
round(cormat,2)
sns.heatmap(cormat);
```
